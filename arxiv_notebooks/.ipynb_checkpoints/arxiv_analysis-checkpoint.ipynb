```python
%matplotlib inline
import pandas as pd
from datetime import datetime as dt
from matplotlib import pyplot as plt
```

```python
df = pd.read_json("data/cs_arxiv.json")
```

```python
df["quantum_appears"] = df.summary.apply(lambda x: 'quantum' in x)
```

```python
df["year"] = df["created"].apply(lambda x: dt.fromtimestamp(x/1000).year)
```

```python
data = [(year, sum(_df.quantum_appears)/len(_df)) for year, _df in df.groupby("year")]
years = [year for (year, frac) in data]
frac = [frac for (year, frac) in data]

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(years, frac)
ax.set_ylabel("Fraction of all CS papers in year\ncontaining 'quantum' in the summary\n", fontsize=16)
_ = ax.set_xlabel("\nYear", fontsize=16)
```
